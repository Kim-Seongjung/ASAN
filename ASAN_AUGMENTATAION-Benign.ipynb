```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def aug(np_img ,crop_img_row , crop_img_col , label):
    """
    np_img must 4D 
    np_img shape : n , row , col , color_ch
    
    """
    n_img,img_row,img_col,color_ch=np.shape(np_img)
    n_ret_img = n_img*(img_row - crop_img_row) * (img_col - crop_img_col)*3
    ret_images = np.zeros([n_ret_img ,crop_img_row , crop_img_col,color_ch])
    len_label= np.shape(label)[1]
    ret_labels =  np.zeros([n_ret_img  ,len_label])
    #print "n_augmented image size : " , n_ret_img 
    #print "n classes :", len_label
    #copy label in factor by 2014
    count=0
    if len(np.shape(np_img))==2:
        np_img=np.reshape(np_img , newshape = [np.shape(np_img)[0] , img_row , img_col ,color_ch])
        print np.shape(np_img)
    for n  in range(0,n_img):
        for r in range(img_row - crop_img_row):
            for c in range(img_col - crop_img_col):
                cropped_img = np_img[n, r:crop_img_row +r , c:crop_img_col+c ,: ]
                ret_images[count*3,:,:,:]=cropped_img  
                ret_images[(count*3+1) , :,:,:] =np.fliplr(cropped_img )
                ret_images[(count*3+2) , :,:,:] =np.flipud(cropped_img )
                
                ret_labels[count*3 , : ] = label[n,:]
                ret_labels[count*3+1 , : ] = label[n,:]                
                ret_labels[count*3+2 , : ] = label[n,:]
                count+=1
    return ret_images ,ret_labels
```

```python
def crawl_folder(folder_path): #search_str 찾고자 하는 string 
    search_path_list=[]
    fileList = os.walk(folder_path).next()[2]
    subFolder_list = os.walk(folder_path).next()[1]
    if(len(fileList)!=0):
        for j in range(len(fileList)):
            search_path_list.append(folder_path+'/'+fileList[j])
    if len(subFolder_list)==0:
        return search_path_list
    else: 
        for i in range(len(subFolder_list)):
            search_path_list.extend(crawl_folder(folder_path+'/'+subFolder_list[i] ))
        return search_path_list
```

# save Augmentated Benign Image

```python
mal_crop_folder = '../Mal_crop/'
benign_crop_folder = '../Benign_crop/'
mal_paths=crawl_folder(mal_crop_folder)
benign_paths=crawl_folder(benign_crop_folder)
new_mal_path=[]
new_benign_path=[]
for i,ele in enumerate(mal_paths):
    if '.npy' in ele:
        new_mal_path.append(ele)
mal_paths = new_mal_path
for i,ele in enumerate(benign_paths):
    if '.npy' in ele:
        new_benign_path.append(ele)
benign_paths = new_benign_path
```

```python
save_path= '/home/user01/hdd1/ASAN_Benign_crop_aug/'
saved_file_path=crawl_folder(save_path)
saved_name=[]
for i,path in enumerate(saved_file_path):
    name=path.split('/')[-1].split('.')[0]
    saved_name.append(name)
print saved_name
```

```python
save_path= '/home/user01/hdd1/ASAN_Benign_crop_aug/'
for i,path in enumerate(benign_paths):
    name=path.split('/')[-1].split('.')[0]
    if name+'_img' in saved_name:
        print 'this file is alerady existed' , name
        continue;
    name=path.split('/')[-1].split('.')[0]
    img=np.load(path)
    lab=np.zeros([1,2])
    lab[0,0]=1
    if len(np.shape(img))==2:
        img=np.expand_dims(img , axis=0)
        img=np.expand_dims(img , axis=4)
    aug_imgs , aug_labs=aug(img , 64,64,lab)
    np.save(save_path +name+'_img' , aug_imgs)
    np.save(save_path +name+'_lab' , aug_labs)
    print name,'is augmented!'
```
